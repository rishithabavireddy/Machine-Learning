```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])


squared = arr ** 2


mean_val = np.mean(arr)
std_val = np.std(arr)

print("Original array:", arr)
print("Squared array:", squared)
print("Mean:", mean_val)
print("Standard Deviation:", std_val)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
result = np.dot(A, B)
print(result)
```

```python
rand_arr = np.random.rand(3, 3)  # 3x3 matrix of random floats
print(rand_arr)
```

```python
arr = np.arange(12)
reshaped = arr.reshape(3, 4)
print(reshaped)
```

```python

arr = np.array([10, 20, 30, 40, 50])
filtered = arr[arr > 25]
print(filtered)
```

```python
arr = np.array([1, 2, 3])
result = arr + 5
print(result)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
stacked = np.vstack((a, b))
print(stacked)
```

```python
matrix = np.array([[1, 2], [3, 4], [5, 6]])
print(matrix.T)
```

```python
arr = np.array([1, 2, 2, 3, 4, 4, 5])
unique_vals = np.unique(arr)
print(unique_vals)
```
