```python
import pandas as pd
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame({'Name': ['Alice','Bob','Charlie'], 'Age':[25,30,35]})
filtered = df[df['Age'] > 28]
print(filtered)
```

```python
df = pd.DataFrame({'Dept':['HR','IT','HR','IT'], 'Salary':[3000,4000,3500,4500]})
grouped = df.groupby('Dept')['Salary'].mean()
print(grouped)
```

```python
df = pd.DataFrame({'Name':['Alice','Bob','Charlie'], 'Age':[25,30,35]})
sorted_df = df.sort_values(by='Age', ascending=False)
print(sorted_df)
```

```python
df = pd.DataFrame({'Name':['Alice','Bob'], 'Age':[25,30]})
df['Age_in_5_years'] = df['Age'] + 5
print(df)
```

```python

df = pd.DataFrame({'Name':['Alice','Bob'], 'Age':[25,30], 'City':['NY','LA']})
df = df.drop(columns=['City'])
print(df)
```

```python
df1 = pd.DataFrame({'ID':[1,2], 'Name':['Alice','Bob']})
df2 = pd.DataFrame({'ID':[1,2], 'Salary':[3000,4000]})
merged = pd.merge(df1, df2, on='ID')
print(merged)
```

```python
df = pd.DataFrame({'Dept':['HR','IT','HR','IT'], 'Salary':[3000,4000,3500,4500]})
pivot = df.pivot_table(values='Salary', index='Dept', aggfunc='mean')
print(pivot)
```
